```python
import numpy as np
import pandas as pd
import csv
import datetime
import time
import random
import os
import matplotlib.pyplot as plt
```

```python
def Check_status(file_path):
    if os.path.exists(file_path):
        os.remove(file_path)
        print("exist")
    return True
```

```python
def average_pmdata():
    sum = 0
    for x in range(20):
        x = random.randint(0,10)
        sum = x + sum
    return round(sum/15, 2)
```

```python
def data_access(count, input_id, pmdata, now):
    a = np.array([count, input_id, now, pmdata], dtype = str)
    return a
```

```python
id = [90, 91, 92, 93]
data_total = []
for x in range(30):
    for input_id in id:
        pmdata = average_pmdata()
        # print(pmdata)
        current = datetime.datetime.now()
        now = str(current.hour) + ":" + str(current.minute) + ":" + str(current.second)
        row_data = data_access(x, input_id, pmdata, now)
        data_total.append(row_data)  
    time.sleep(2)
df = pd.DataFrame(data_total, columns = ["count", "id", "time", "pmdata"])
# print(df)
```

```python
df2 = pd.DataFrame({'A': np.array(df[0:len(df.index):4].pmdata), 
                    'B': np.array(df[1:len(df.index):4].pmdata), 
                    'C': np.array(df[2:len(df.index):4].pmdata), 
                    'D': np.array(df[3:len(df.index):4].pmdata)}, index = df[0:len(df.index):4].time)
print(df2)
df2 = df2.astype(float)
df2.plot()
plt.show()
```
